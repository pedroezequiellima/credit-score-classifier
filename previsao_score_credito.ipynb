Projeto de Previsão com Machine Learning


**Exploração de Dados**

```python
#Passo a Passo do projeto
# Passso 0: Entender o dasafio da empresa
# Passo 1: Importar a base de dados

import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)

print(tabela.info())
```

**Pré-processamento**

```python
# Passo 2: tratar e preparar a base de dados para a inteligencia artificial
# Quero prever a coluna 'score de credito'
# profissao, mix de credito e a comportamento de pagamento: Todas sao colunas de credito

from sklearn.preprocessing import LabelEncoder

# profissao: cientista vai ser 1, marceneiro=2, jogador=3
codificador_profissao = LabelEncoder()
tabela['profissao'] = codificador_profissao.fit_transform(tabela["profissao"])


codificador_mix = LabelEncoder()
tabela['mix_credito'] = codificador_mix.fit_transform(tabela["mix_credito"])

codificador_comportamento = LabelEncoder()
tabela['comportamento_pagamento'] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])


display(tabela)
```

**Treinamento do Modelo**

```python
# como fuciona o processo para criar uma inteligencia artificial(modelo)

# X es quem eu vou usar para fazer a previssao
x = tabela.drop(columns=["score_credito", "id_cliente"])

# Y es quem eu vou prever
y = tabela["score_credito"]

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

**Criar o modelo de IA**

```python
# Passo 3: Criar o modelo de IA, previsao -> bom, ruim, ok

# Passo 1: Importar a IA(modelo)
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Passo 2: Criar a IA(modelo)
modelo_arvore_decisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Passo 3: Treinar a IA(modelo)
modelo_arvore_decisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

**Avaliando a porcentagem de acertos dos modelos com a base de dados para treino.**

```python
# Passo 4: Avaliar modelos e escolher o melhor

# Previsao
previsao_arvore = modelo_arvore_decisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# Comparar a previsao com Y Teste:
from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvore))
display(accuracy_score(y_teste, previsao_knn))
```

**Colocando o modelo para prever o score de credito em uma nova base de dados!**

```python
# Passo 5: Fazer novas previsoes

# Melhor modelo = Arvore de decisao

# Importar os novos clientes
novos_clientes = pd.read_csv("novos_clientes.csv")

novos_clientes['profissao'] = codificador_profissao.fit_transform(novos_clientes["profissao"])
novos_clientes['mix_credito'] = codificador_mix.fit_transform(novos_clientes["mix_credito"])
novos_clientes['comportamento_pagamento'] = codificador_comportamento.fit_transform(novos_clientes["comportamento_pagamento"])

previsao = modelo_arvore_decisao.predict(novos_clientes)
novos_clientes["previsao"] = previsao

display(novos_clientes)
display(previsao)

```
